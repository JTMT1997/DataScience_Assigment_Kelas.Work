```python
#Check Na
total = house.isnull().sum().sort_values(ascending=False)
percent = (house.isnull().sum()/house.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total,percent],axis=1, keys=['Total','Percent'])
missing_data.head(20)
```

```python
#Drop NaN
house = house.dropna(axis=1)
house.head()
```

```python
#Memisahkan Data Numeric dan Categorical
#Mendapatkan data numeric
num_cols = house._get_numeric_data().columns
house_numeric = house[num_cols]
house_numeric.head()
```

```python
#Mendapatkan data kategorical
house_categorical = house.drop(num_cols, axis=1)
house_categorical.head()
```

```python
#Menyimpan Data ke dalam file (csv)
house_numeric.to_csv('house_numeric.csv')
```

```python
house_categorical.to_csv('house_categorical.csv')
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
house_categorical = pd.read_csv('house_categorical.csv',index_col=0)
house_numerical = pd.read_csv('house_numeric.csv',index_col=0)
```

```python
sns.barplot(x=house_categorical['Utilities'].value_counts().index,y=house_categorical['Utilities'].value_counts(normalize = True))
plt.title ('Bar Chart')
plt.xlabel ('Utilities')
plt.ylabel ('Proportion')
plt.show()
```

```python
x = house_numerical['SalePrice']
sns.distplot(x, axlabel = x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label= "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.title('Histogram')
plt.show()
```

```python
#Transformasi log10 pada variabel SalePrice
x = np.log10(house_numerical['SalePrice'])
sns.distplot(x, axlabel = x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label= "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.title('Histogram')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
house_categorical = pd.read_csv('house_categorical.csv',index_col=0)
house_numerical = pd.read_csv('house_numeric.csv',index_col=0)
```

**Scatter Plot**

```python
ax = sns.scatterplot(y='SalePrice', x='GrLivArea', data=house_numerical)
plt.title('Linear graph')
plt.ylabel('SalePrice')
plt.xlabel('GrLiveArea')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import seaborn as sns
%matplotlib inline
ax = sns.scatterplot(y='SalePrice', x='GarageArea', data=house_numerical)
plt.title('Linear graph')
plt.ylabel('SalePrice')
plt.xlabel('GarageArea')
plt.show()
```

```python
#Filtering GarageArea value > 0
filtered_df = house_numerical[(house_numerical['GarageArea']>0)]
filtered_df
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import seaborn as sns
%matplotlib inline
ax = sns.scatterplot(y='SalePrice', x='GarageArea', data=filtered_df)
plt.title('Linear graph')
plt.ylabel('SalePrice')
plt.xlabel('GarageArea')
plt.show()
```

```python
#Menghitung nilai korelasi antara variabel GrlivArea dengan SalePrice

r_GrLiveArea_SalePrice = house_numerical['GrLivArea'].corr(house_numerical['SalePrice'])
print(r_GrLiveArea_SalePrice)
```

```python
#Menghitung nilai korelasi antara variabel GrlivArea dengan SalePrice

r_GarageArea_SalePrice = house_numerical['GarageArea'].corr(house_numerical['SalePrice'])
print(r_GarageArea_SalePrice)
```
